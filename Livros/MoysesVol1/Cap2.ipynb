```python
import numpy as np
import matplotlib.pyplot as plt

#Definicoes padrao dos grafico
plt.rcParams['figure.figsize'] = [10, 10] 
plt.rcParams['axes.axisbelow'] = True
plt.rcParams['axes.grid'] = True
plt.rcParams['axes.labelsize'] = 15
```

## Exercício 6

O exercício pede a velocidade e a posicao da particula em funcao do tempo

```python
#Conjunto de pontos para o tempo
t = np.linspace(0, 100, 100)

b = 0.5 #Dado no enunciado

v = 0.5 * b * t**2.
x = 1./6. * b * t**3.
```

```python
fig, ax = plt.subplots(2, 1, sharex=True, sharey=False)
plt.subplots_adjust(wspace=0.0,hspace=0.0)


ax[0].plot(t, v, color='darkorange')
ax[1].plot(t, x, color='royalblue')

ax[0].set_ylabel('Velocidade [m/s]')
ax[1].set_ylabel('Posição [m]')

ax[1].set_xlabel('Tempo [s]')
```
